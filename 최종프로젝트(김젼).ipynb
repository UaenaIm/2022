#당첨자 랜덤뽑기<발전편>


```python
import random
import time
import csv
data = csv.reader(open('name.csv'))
next(data)
myname = []
for row in data :
  myname.append(row[1])

prize = ['사탕 n개 가져가기', '사탕 2개 뽑아서 친구에게 주기', '노래 하고 사탕 5개 또는 사탕 안 받기', '한 주먹 뽑기', '젓가락으로 집어 가기', '친구가 한 주먹 뽑아주기', '친구랑 가위바위보 해서 이기면 사탕 2개', '뒷 번호 친구 칭찬 3개 하고 사탕 3개 받기']

print('당첨자는')
time.sleep(5)
print(random.choice(myname)+ ' 입니다!')

time.sleep(3)

while True:
    n = random.randint(1,4)
    m = random.randint(1,25)
    prize_result = random.choices(prize, weights=(30, 15, 15, 5, 10, 5, 10, 10), k=1)

    if prize_result[0] == prize[0] :
      print('당첨 결과는..!!')
      print( f'{prize_result} 입니다!')
      time.sleep(2.5)
      print('당첨자가 가지고 갈 사탕의 개수는')
      print( str(n) + '개 입니다')     
      break

    elif prize_result[0] == prize[1] :
      print('당첨 결과는..!!')
      print(f'{prize_result} 입니다!')
      time.sleep(2.5)
      print('당첨자의 사탕을 받을 친구는!')
      print(random.choice(myname) + ' 입니다!')
      break


    elif prize_result[0] == prize[5] :
      print('당첨 결과는..!!')
      print(f'{prize_result} 입니다!')
      time.sleep(2.5)
      print('당첨자의 사탕을 뽑아줄 친구는!')
      print(random.choice(myname) + ' 입니다!')
      break

 
    elif prize_result[0] == prize[6] :
      print('당첨 결과는..!!')
      print(f'{prize_result} 입니다!')
      time.sleep(2.5)
      print('당첨자와 가위바위보를 할 친구는!')
      print(random.choice(myname) + ' 입니다!')
      break     

    else : 
      print('당첨 결과는..!!')
      print(f'{prize_result} 입니다!')
      break
```
